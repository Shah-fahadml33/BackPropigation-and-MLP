```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_train=np.random.rand(1000,3)*10
true_w = np.array([1.0, 0.2, 0.2])   
true_b = 1.0

y_train = x_train @ true_w + true_b + np.random.randn(1000)
```

```python

lr=0.001
epochs=2000

```

```python
def Mini_BGD(x_train,y_train,lr,batch_size,epochs):
    m,n=x_train.shape
    w=np.zeros(n)
    b=0.0
    
    losses=[]
    
    for epoch in range(epochs):
        idx=np.random.permutation(m)
        x_shuffle=x_train[idx]
        y_shuffled=y_train[idx]
    
        for i in range(0,m,batch_size):
            x_batch=x_shuffle[i:i+batch_size]
            y_batch=y_shuffled[i:i+batch_size]

            B=y_batch.shape[0]

            y_hat=np.dot(x_batch,w)+b

            error=y_hat-y_batch

            dw=(1/B)*np.dot(x_batch.T,error)
            db=(1/B)*np.sum(error)

            w-=lr*dw
            b-=lr*db

        cost=((np.dot(x_train,w)+b)-y_train)
        loss=(1/(2*m))*np.sum(cost**2)
        losses.append(loss)

    return w,b,losses
```

```python
learned_w,learned_b,losses_l=Mini_BGD(x_train,y_train,lr,batch_size=32,epochs=2000)
```
