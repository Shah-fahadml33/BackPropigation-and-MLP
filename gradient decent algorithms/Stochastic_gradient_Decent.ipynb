```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_train=np.random.rand(100,4)*10
true_w=np.array([1,0.1,0.02,5])
true_b=1.0
y_train=x_train@true_w+true_b*np.random.randn(100)

w=np.zeros(4)
b=0.0
lr=0.001
epochs=2000

losses=[]

for epoch in range(epochs):
    indices=np.random.permutation(len(x_train))
    for idx in indices:
        x_i=x_train[idx]
        y_i=y_train[idx]

        pred=np.dot(w,x_i)+b
        error=pred-y_i
        dw=np.dot(error,x_i)
        db=error

        w-=lr*dw
        b-=lr*db

        loss=(1/(1*len(x_train)))*(error**2)
        losses.append(loss)
    if epoch %100==0:
              print(f"Epoch :{epoch} |Loss :{loss:.4f}") 
```

```python
plt.plot(losses)
```
