```python
import numpy as np
import matplotlib.pyplot as plt
```

```python


def batch_gradient_descent(X, y, w_init, b_init, alpha, epochs):

    m = len(y)
    w = w_init
    b = b_init
    cost_history = []
    
    for i in range(epochs):
        # 1. Calculate predictions (The Hypothesis)
        predictions = w * X + b
        
        # 2. Calculate the error (Prediction - Actual)
        error = predictions - y
        
        # 3. Compute the gradients (Matching our exact mathematical derivation)
        # np.sum() aggregates across all m examples instantly
        dw = (1/m) * np.sum(error * X) 
        db = (1/m) * np.sum(error)     
        
        # 4. Update the parameters simultaneously
        w = w - alpha * dw
        b = b - alpha * db
        
        # 5. Calculate and record the cost (Mean Squared Error)
        cost = (1 / (2 * m)) * np.sum(error ** 2)
        cost_history.append(cost)
        
        # Optional: Print progress every 10% of epochs
        if i % (epochs // 10) == 0:
            print(f"Epoch {i:4d}: Cost = {cost:.4f}, w = {w:.4f}, b = {b:.4f}")
            
    return w, b, cost_history
```

```python
np.random.seed(42)
x_train=np.random.rand(100)*10
y=2*x_train+1*np.random.rand(100)
w=np.random.random()
b=np.random.random()
lr=0.01
epochs=1000
```

```python
w,b,cost=batch_gradient_descent(x_train,y,w,b,lr,epochs)
```

```python
plt.plot(cost)
plt.xlabel("Epochs")
plt.ylabel("Loss")
```
